```python
dafr = dafr.drop(['Booking_ID', 'avg_price_per_room'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder

for c_l in list(dafr.select_dtypes(include=['object']).columns):
    dafr[c_l] = LabelEncoder().fit_transform(dafr[c_l])
    
dafr.info()
dafr.head(10)
```

```python
# for knn
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_scaled = scaler1.fit_transform(dafr)
dafr[dafr.columns] = X_scaled
dafr.head(10)
```

```python
dafr.info()
```

```python
from sklearn.model_selection import train_test_split

X = dafr.drop(['room_type_reserved'], axis = 1)
y = dafr['room_type_reserved']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
scaler2.fit(X_train)
```

```python
scaled_X_train = scaler2.transform(X_train) 
scaled_X_test = scaler2.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

```python
y_train_tr = LabelEncoder().fit_transform(y_train)
knn.fit(scaled_X_train, y_train_tr)
```

```python
pred = knn.predict(scaled_X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score

y_test_tr = LabelEncoder().fit_transform(y_test)
print(confusion_matrix(y_test_tr,pred))
```

```python
print(classification_report(y_test_tr, pred, zero_division=0.0))
```

```python
print(accuracy_score(y_test_tr,pred))
```

```python
import numpy as np

error_rate = []

for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(scaled_X_train,y_train_tr)
    pred_i = knn.predict(scaled_X_test)
    error_rate.append(np.mean(pred_i != y_test_tr))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,30), error_rate, marker='o', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.grid()
```

```python
knn = KNeighborsClassifier(n_neighbors=13)

knn.fit(scaled_X_train,y_train_tr)
pred = knn.predict(scaled_X_test)

print(confusion_matrix(y_test_tr,pred))
print()
print(classification_report(y_test_tr, pred, zero_division=0.0))
```
